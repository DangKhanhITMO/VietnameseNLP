```python
#1. Получить текстовые данные в википедии по заданному ключевому слову с поиском на вьетнамском языке 
import wikipedia
def get_documentFromWiki_bykeyword(keyword):
    language = "vi"
    wikipedia.set_lang(language)
    result = wikipedia.search(keyword)
    page = wikipedia.page(result[0])
    document= page.content
    return document
```

```python
#2.Создание корпус слов, взятых из данных в Википедии, по нескольким ключевым словам поиска. 
#В этой работе собраны данные из 15 статей в Википедии.
def get_corpus(arrayKeyword):
    documentCorpus=""
    for keyword in range(0,len(arrayKeyword)):
        documentCorpus+=get_documentFromWiki_bykeyword(arrayKeyword[keyword])
    return documentCorpus
```

```python
#3. Первичная обработка
def preprocess(inputdocument):
    outputDocument=text_preprocess(inputdocument)
    return outputDocument
```

```python
#4. Переобразование на тип множества для удобнее обработки
def toSetCorpus(inputdocument):
    outputSet=set()
    for elem in inputdocument.split():
        outputSet.add(elem)
    return outputSet
```

```python
#5. Удаление окончания слов
def deleteStopWord(inputdocument):
    outputdocument=inputdocument-setStopWord
    return outputdocument
```

```python
#6. Формирование корпуса слов типа Directory in python и считать частуту слов
def toDirector(inputSetWord, inputCorpusDocument):
    outputDirector=dict.fromkeys(inputSetWord,0)
    for word in inputCorpusDocument.split():
        if (word in inputSetWord):
            outputDirector[word]+=1
    sortOutput={}
    sorted_keys = sorted(outputDirector, key=outputDirector.get,reverse=True)
    for w in sorted_keys:
        sortOutput[w] = outputDirector[w]
    return (sortOutput)
```
